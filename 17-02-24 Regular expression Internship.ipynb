```python
# Extracting digit from string:
Write a regular expression to search digit inside a string
```

```python
# Find digit

pattern = '\d+'
text = "There are 12 3 appels and also 450 oranges."
matches = re.findall(pattern,text)
print(matches)
```

```python
pattern = '123|456'
text = 'There are 1234 appels and also 456 oranges.'
matches = re.findall(pattern,text)
print(matches)
```

```python
# find only characters
```

```python
# Except digit only characters.

pattern = "\D+"
text = "There are 12 appels and 456 oranges."
matches = re.findall(pattern,text)
print(matches)
```

```python
pattern = "[0-9]+"
text = "There are 12 3 appels and 456 oranges."
matches = re.findall(pattern,text)
print(matches)
```

```python
pattern = "[0-9]+"
matches = re.findall(pattern,'abc1235@#8521_0')
print(matches)
```

```python
# Extracting digit from a list of string
```

```python
price = ['Apple cost Rs 50','Rs 60 for each pineapple 120','Rs 150 per watermilin']
for msg in price:
    matches = re.findall('\d+',msg)
    print(matches)
```
